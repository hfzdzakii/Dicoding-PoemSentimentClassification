```python
import pandas as pd
import os
```

Just doing a merge things. 150 parts into 1 single file!

```python
def merging():
    arr = []
    for file in sorted(os.listdir(f'./poem')):
        if file.endswith('.csv'):
            print(f"Merge {file}")
            poem = pd.read_csv(f'./poem/{file}')
            arr.append(poem['0'])
    if arr:
        ser = pd.concat(arr, ignore_index=True, axis=0)
        ser.dropna(inplace=True)
        ser.drop_duplicates(inplace=True)
        ser.name = 'poem'
        print("Save to csv")
        ser.to_csv(f"./dataset/poem_dataset.csv", index=False)
    else:
        print("No array")
            
```

```python
merging()
```
